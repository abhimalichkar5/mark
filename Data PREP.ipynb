```python
import pandas as pd
```

```python
# Dataset pre-processing

AMZN = pd.read_csv("./intraday/AMZN.csv")

# Drop first 2 rows
AMZN.drop(AMZN.head(2).index, inplace=True)

# Drop the row with #NAME?
# AMZN.drop(3, inplace=True)

# Change the headers
new_head = AMZN.iloc[0]
AMZN = AMZN[1:]
AMZN.columns = new_head

# Keep just the first 8 columns - concerned with Trade data
AMZN = AMZN.iloc[:, :8]

# Delete all rows where Dates is NaN
AMZN = AMZN[AMZN['Dates'].notna()]

# Remove the 2 in top left
AMZN.columns.name = None

AMZN['Dates'] = pd.to_datetime(AMZN['Dates'])

AMZN.set_index('Dates', inplace=True)

# Get only market open and noon data
mask = ((AMZN.index.hour == 9) & (AMZN.index.minute == 30)) | ((AMZN.index.hour == 12) & (AMZN.index.minute == 00))
AMZN = AMZN[mask]

AMZN.to_csv("./data/AMZN_processed.csv")
```

```python
TSLA = pd.read_csv("./intraday/TSLA.csv")

# Drop first 2 rows
TSLA.drop(TSLA.head(2).index, inplace=True)

# Change the headers
new_head = TSLA.iloc[0]
TSLA = TSLA[1:]
TSLA.columns = new_head

# Keep just the first 8 columns - concerned with Trade data
TSLA = TSLA.iloc[:, :8]

# Delete all rows where Dates is NaN
TSLA = TSLA[TSLA['Dates'].notna()]

# Delete all rows where Dates is NaN
TSLA = TSLA[TSLA['Dates'].notna()]

# Remove the 2 in top left
TSLA.columns.name = None

TSLA['Dates'] = pd.to_datetime(TSLA['Dates'])

TSLA.set_index('Dates', inplace=True)

mask = ((TSLA.index.hour == 9) & (TSLA.index.minute == 30)) | ((TSLA.index.hour == 12) & (TSLA.index.minute == 00))
TSLA = TSLA[mask]

TSLA.to_csv("./data/TSLA_processed.csv")
```

```python
# Dataset pre-processing

NFLX = pd.read_csv("./intraday/NFLX.csv")

# Drop first 2 rows
NFLX.drop(NFLX.head(2).index, inplace=True)

# Drop the row with #NAME?
# NFLX.drop(3, inplace=True)

# Change the headers
new_head = NFLX.iloc[0]
NFLX = NFLX[1:]
NFLX.columns = new_head

# Keep just the first 8 columns - concerned with Trade data
NFLX = NFLX.iloc[:, :8]

# Delete all rows where Dates is NaN
NFLX = NFLX[NFLX['Dates'].notna()]

# Remove the 2 in top left
NFLX.columns.name = None

NFLX['Dates'] = pd.to_datetime(NFLX['Dates'])

NFLX.set_index('Dates', inplace=True)

mask = ((NFLX.index.hour == 9) & (NFLX.index.minute == 30)) | ((NFLX.index.hour == 12) & (NFLX.index.minute == 00))
NFLX = NFLX[mask]

NFLX.to_csv("./data/NFLX_processed.csv")
```

```python
AMZN = pd.read_csv("./data/AMZN_processed.csv")
NFLX = pd.read_csv("./data/NFLX_processed.csv")
TSLA = pd.read_csv("./data/TSLA_processed.csv")
```

```python
# Since we cannot exactly buy at the open or closing price,
# we're taking the average price of that time period


# AMZN['Average'] = (AMZN['Open'] + AMZN['Close'])/2
# NFLX['Average'] = (NFLX['Open'] + NFLX['Close'])/2
# TSLA['Average'] = (TSLA['Open'] + TSLA['Close'])/2
```

```python
# Getting the data for just the dates common in all 3 stocks

common_dates_AMZN = AMZN['Dates'].isin(TSLA['Dates']) & AMZN['Dates'].isin(NFLX['Dates'])
common_dates_NFLX = NFLX['Dates'].isin(TSLA['Dates']) & NFLX['Dates'].isin(AMZN['Dates'])
common_dates_TSLA = TSLA['Dates'].isin(AMZN['Dates']) & TSLA['Dates'].isin(NFLX['Dates'])

AMZN_filtered = AMZN[common_dates_AMZN]
NFLX_filtered = NFLX[common_dates_NFLX]
TSLA_filtered = TSLA[common_dates_TSLA]
```

```python
print(len(AMZN_filtered), len(TSLA_filtered), len(NFLX_filtered))
```

```python
returns_dict = {'Dates': [], 'AMZN': [], 'NFLX': [], 'TSLA': []}
returns = pd.DataFrame(returns_dict)
```

```python
returns['Dates'] = pd.to_datetime(AMZN_filtered['Dates'][::2]).dt.date

returns['AMZN'] = (AMZN_filtered['Close'][1::2].values - AMZN_filtered['Close'][::2].values) / (AMZN_filtered['Close'][::2].values)
returns['NFLX'] = (NFLX_filtered['Close'][1::2].values - NFLX_filtered['Close'][::2].values) / (NFLX_filtered['Close'][::2].values)
returns['TSLA'] = (TSLA_filtered['Close'][1::2].values - TSLA_filtered['Close'][::2].values) / (TSLA_filtered['Close'][::2].values)
```
